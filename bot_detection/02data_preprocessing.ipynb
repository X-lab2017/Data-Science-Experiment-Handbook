```python
data = raw_data[columns]
```

### 2. 缺失值处理

```python
# 查看各列的缺失值
```

```python
#把 type 转化成 boolean 变量 

print(data['type'].value_counts())

data['type'] = data['type'].apply(lambda x: x == 'User')

data['type'].value_counts()
```

```python
# 填充 hireable 的缺失值

print(data['hireable'].value_counts())

data['hireable'].fillna(False, inplace=True)

print(data['hireable'].value_counts())

```

```python
data['location_exists'] = data['location'].notna()
data['company_exists'] = data['company'].notna()
data['blog'] = data['blog'].notna()
data['location_exists'].value_counts()
```

```python
# 将created_at、updated_at转为时间戳

from datetime import datetime
def convert_to_timestamp(time_str):
    if not pd.isna(time_str):
        return datetime.strptime(time_str, '%Y-%m-%dT%H:%M:%SZ').timestamp()
    else:
        return None
    
data['created_at'] = data['created_at'].apply(convert_to_timestamp)
data['updated_at'] = data['updated_at'].apply(convert_to_timestamp)

data[['created_at', 'updated_at']]
```
